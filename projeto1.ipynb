## 3 - Conside os dados da Iris e mostre a distribuição de probabilidades de cada uma das variáveis após a normalização e padronização.

```
# Bibliotecas utilizadas:

import pandas as pd
import pandas.util.testing as tm
import numpy as np
from sklearn.preprocessing import MinMaxScaler # biblioteca para normalizar 
import matplotlib.pyplot as plt
from scipy.stats import shapiro
import seaborn as sns
import pandas.util.testing as tm
from sklearn.preprocessing import StandardScaler
```

### Normalização

Lendo o *dataset* e verificando se a leitura foi realizada corretamente:

```
data = pd.read_csv('/content/aula2_data/iris.csv', header=(0))

data.head(5)
```

Normalizando no intervalo [0, 1]:

```
data1 = data.iloc[:, [0,1,2,3]] # Pegando somente as 4 primeiras colunas
data1 = np.array(data1, dtype = float) # Transformando em array do numpy

# Prepara a função para transformar (normalizar) os dados
scaler = MinMaxScaler(feature_range = (0, 1))

# Realizando a normalização e colocando em um array
data_norm = scaler.fit_transform(data1)

# Transformando novamente em dataframe
data_norm = pd.DataFrame(data=data_norm, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])

data_norm.head(15)
```

Exibindo a distribuição de probabilidades de cada uma das variáveis após a normalização:

```
aux = ['sepal length', 'sepal width', 'petal length', 'petal width']

plt.figure(figsize=(15,18))

for i in range(len(aux)):
  plt.subplot(2, 2, i + 1)
  
  values, bins, _ = plt.hist(data_norm.iloc[:, i], color = 'darkblue', density = 'true', bins = 'doane', edgecolor='black')
  plt.title('Histogram of ' + aux[i])
  plt.xlabel(aux[i].lower().capitalize())
  plt.ylabel('Density')

  # Conferindo se as áreas são 1
  print("Área do histograma ", i + 1, ": ", round(sum(np.diff(bins)*values), 4))

print()
plt.show()
```

Nota-se, para as duas primeiras variáveis (*sepal lenght* e *sepal width*), uma leve simetria em seus histogramas. Para as variáveis *petal length* e *petal width* percebe-se que, visualmente, os dados são oriundos de distrições bimodais. 

Apresenta-se, juntamente com o histograma, uma curva de suavização dos dados, que nada mais é do que uma estimativa não paramétrica para a função densidade de probabilidade associada a variável aleatória em questão. O método de estimação consiste em fazer uma suavização de cada ponto $X_i$ e depois somá-las; em outras palavras, pondera-se a distância de cada observação em relação ao núcleo. Tal estimativa pode ser dada da seguinte forma:

$$\hat{f}_{h}(x)=\frac{1}{n h} \sum_{i=1}^{n} K\left(\frac{x-x_{i}}{h}\right)$$

Em que $K(•)$ é um kernel qualquer. Mais usualmente,

$$K(x)=\frac{1}{\sqrt{2 \pi}} e^{-\frac{x^{2}}{2}}$$

que é denominado como núcleo de Gauss ou função de Gauss. 


```
# Distribuição não paramétrica + histograma das probabilidades

aux = ['sepal length', 'sepal width', 'petal length', 'petal width']

plt.figure(figsize=(15,18))

for i in range(len(aux)):
  plt.subplot(2, 2, i + 1)

  sns.distplot(data_norm.iloc[:,i], hist=True, kde=True, 
             bins='doane', color = 'darkblue',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
  
  plt.title('Histogram of ' + aux[i])
  plt.xlabel(aux[i].lower().capitalize())
  plt.ylabel('Density')

plt.show()
```

Assim como antes, aqui, fica mais evidente por meio da estimativa não paramétrica que os dados dos atributos *sepal lenght* e *sepal width* possuem, possivelmente, uma distribuição simétrica. Ademais, fica nítido que para os outros dois atributos as distribuições de probabilidades a eles associadas são bimodais.

A seguir, verifica-se se os dados são normais ou não. Para este fim, utiliza-se do teste de normalidade de Shapiro-Wilk, um dos mais difundidos para tal tarefa.
Ainda, ressalta-se que o nível de significância adotado foi de 5%.

```
# Teste de Normalidade de Shapiro-Wilk

for i in range(4):
  print("Atributo", aux[i], ": ", shapiro(data_norm.iloc[:, i]))
```

Os resultados obtidos podem ser observados na tabela a seguir (construída para melhor/mais fácil observação dos resultados):

  |   Atributos  | Estatística do Teste | Valor-p |
  |:------------:|:--------------------:|:-------:|
  | sepal length |       0.9761         | 0.0102  |
  | sepal width  |       0.9838         | 0.0752  |
  | petal length |       0.8764         |    0    | 
  | petal width  |       0.9026         |    0    |

Com base nos resultados, a nível de significância de 5% é possível concluir que para os dois primeiros atributos (*sepal lenght* e *sepal width*), após a normalização, não existem indícios que nos levem a crer que os dados não são normais. O mesmo não ocorre para o terceiro e o quarto atributo (*petal lenght* e *petal widht*), onde rejeita-se a hipótese nula de que os dados são provenientes de uma distribuição normal. 

Como nem todos os atributos apresentam indícios de serem normalmente distribuídos, na seleção do número de bins, opta-se pelo método *doane*, que promete ser uma versão melhorada do regra de *sturges*, melhor aplicável para *datasets* não normalmente distribuídos.

## 4 - Reamostre os dados da Iris e selecione 10 elementos em cada classe.

```
#Bibliotecas utilizadas

import pandas as pd
import numpy as np
import random

random.seed(19)
```

Carregando os dados e verificando se os mesmos foram carregados da maneira correta:

```
data = pd.read_csv('/content/aula2_data/iris.csv', header=(0))

data.head(5)
```

Convertendo variáveis nominais em inteiros (no caso, o atributo nome das espécies):


```
data = data.replace('setosa', 0) 
data = data.replace('versicolor', 1)
data = data.replace('virginica', 2)

data.head(5)
```

A seguir, sorteamos aleatoriamente 10 elementos de cada classe:

```
N = 10  # Número de sorteios /  Tamanho da amostra

# Pegando apenas a espécie de cada registro
classes = data[data.columns[-1]]
# Removendo repetições: ao final teremos apenas as classes Setosa, Versicolor e Vignínica 
cl = np.unique(classes)

# Criando um vetor com todas as informações do dataframe
x = np.array(data, dtype = 'float')

newX = []

cls = np.array(data[data.columns[-1]])

for i in np.arange(0, len(cl)):
    a = np.argwhere(cls == cl[i])

    # Selecionando os índices nos dados de forma aleatória
    inds = np.random.choice(a[:,0], N, replace = False)

    # Concatenando os registros sorteados
    newX.append(x[inds,:])

newX = np.array(newX)

print('Dados obtidos a partir da amostragem:\n')
print(newX)
```

Em seguida, apresenta-se o histograma e a estimativa de densidade de *kernel* do novo conjunto de dados:

```
# Distribuição não paramétrica e histograma das probabilidades

aux = ['sepal length', 'sepal width', 'petal length', 'petal width']

plt.figure(figsize=(15,10))

for i in range(len(aux)):
  plt.subplot(2, 2, i + 1)

  sns.distplot(newX[:,i], hist=True, kde=True, 
             bins='doane', color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
  
  plt.title('Histogram of ' + aux[i])
  plt.xlabel(aux[i].lower().capitalize())
  plt.ylabel('Density')

plt.show()
```

No gráfico acima, após a reamostragem, destaca-se que para os atributos *petal length* e *petal width* houve uma mitigação da característica bimodal antes apresentada. Ainda, nos gráficos dos atributos *sepal length* e *sepal width*, o aspecto simétrico deixou de ser claramente nítido. 

Todas as colocações supracitadas se devem ao reduzido tamanho de amostra considerado. Possivelmente, para maiores números de observações, todos os quatro gráficos seriam semelhantes àqueles já apresentados neste estudo (considerando todas as observações).

Por fim, salva-se o novo conjunto de dados em um csv.

```
# Salvando em dataframes
data_reamost0 = pd.DataFrame(data = newX[0], columns=["sepal_length", "sepal_width", "petal_length", "petal_width", "species"])
data_reamost1 = pd.DataFrame(data = newX[1], columns=["sepal_length", "sepal_width", "petal_length", "petal_width", "species"])
data_reamost2 = pd.DataFrame(data = newX[2], columns=["sepal_length", "sepal_width", "petal_length", "petal_width", "species"])

# Concatenando em um dataframe único
data_reamost = pd.concat([data_reamost0, data_reamost1, data_reamost2])
data_reamost.index = range(0, 30)

# Transformando o species em atributo nominal
data_reamost.species = data_reamost.species.astype(str)
data_reamost = data_reamost.replace("0.0", "setosa")
data_reamost = data_reamost.replace("1.0", 'versicolor')
data_reamost = data_reamost.replace("2.0", 'virginica')


data_reamost.head(35)

data_reamost.to_csv('/content/aula2_data/iris-reamostra.csv')
```

## 6 - Encontre os outliers nos dados das bases Iris e BostonHouse.

```
# Bibliotecas utilizadas
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import warnings
```

### Dataset Iris

Lendo o *dataset* e verificando se tudo ocorreu bem:

```
data = pd.read_csv('/content/aula2_data/iris.csv', header=(0))
data.head(15)
```

Fazendo a análise com base nos boxplots (previamente plotados na seção 5):

```
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

pd.set_option("display.max_rows", None, "display.max_columns", None)

# Setando true ou false para a presença de outlier em quaisquer atributos (exceto nominal)
data['outlier'] = (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1))

data[data.outlier == True]
```

Tem-se, acima, os valores que são considerados *outliers* (4 observações em 150), pelo método do boxplot, para os dados do *dataset* Iris. Essa análise é simplista e preliminar, pois, dentre outras coisas, desconsidera-se o fato de se ter dados multivariados ao analisá-los individualmente, atributo a atributo.

A seguir, identifica-se os *outliers* visualmente em um gráfico para verificar se os registros encontrados são, de fato, *outliers* no scatterplot. Vale ressaltar que as variáveis utilizadas foram escolhidas arbitrariamente.

```
plt.figure(figsize=(20,10))

plt.scatter(x = data.sepal_length, y = data.sepal_width, s = 200,
            c = data.outlier, cmap = 'viridis')

plt.xlabel('sepal_length')
plt.ylabel('sepal_width')

plt.show()
```

Como é possível observar, os *outliers* (representados na cor amarela) são visualmente condizentes com o esperado (esperava-se que os mesmos se encontrassem às margens do gráfico). Ainda, é possível afirmar que os *outliers* estão espacialmente distantes das demais observações.

Por fim, por curiosidade, resolveu-se analisar a correlação entre os quatro atributos existentes através de uma análise gráfica acrescida do coeficiente de correlação de Pearson, como consta a seguir:

```
# Código responsável por incluir o coeficiente de correlação de Pearson no triângulo
#superior do gráfico

def corrdot(*args, **kwargs):
    corr_r = args[0].corr(args[1], 'pearson')
    corr_text = f"{corr_r:2.2f}".replace("0.", ".")
    ax = plt.gca()
    ax.set_axis_off()
    marker_size = abs(corr_r) * 10000
    ax.scatter([.5], [.5], marker_size, [corr_r], alpha=0.6, cmap="coolwarm",
               vmin=-1, vmax=1, transform=ax.transAxes)
    font_size = abs(corr_r) * 40 + 5
    ax.annotate(corr_text, [.5, .5,],  xycoords="axes fraction",
                ha='center', va='center', fontsize=font_size)
```

```
# Gerando o gráfico em si

sns.set(style='white', font_scale=1.6)
g = sns.PairGrid(data.iloc[:,[0, 1, 2, 3]], aspect=1.4, diag_sharey=False).set(ylim=(0, 8))
g.map_lower(sns.regplot, lowess=True, ci=False, line_kws={'color': 'black'})
g.map_diag(sns.distplot, kde_kws={'color': 'black'})
g.map_upper(corrdot)

plt.show()
```

Como é possível observar, existe uma forte correlação entre os atributos *petal width* e *petal length*, *petal length* e *sepal length* e *petal width* e *sepal length*; para os demais, tem-se uma correlação fraca ou moderada.

Lendo os dados e verificando se tudo ocorreu bem:

```
data = pd.read_csv('/content/aula2_data/BostonHousing.csv', header=(0))

data.head(15)
```

Gerando um boxplot para cada atributo:

```
aux = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat', 'medv']

plt.figure(figsize=(20, 50))

for i in range(len(aux)):
  plt.subplot(5, 3, i + 1)

  sns.boxplot(y = aux[i], data=data, color = 'darkblue')
  plt.ylabel(aux[i], fontsize=16)
  plt.yticks(fontsize=16)

plt.show(True)
```

É possível observar que o boxplot relacionado ao atributo *chas* não representa coisa alguma (uma vez que o mesmo se trata de uma *dummy variable*). 

De um modo geral, em cinco atributos não se encontra *outliers*, sendo eles *indus*, *nox*, *age*, *rad* e *tax*. Para todos os demais existe um ou mais *outliers*.

Fazendo a análise com base nos boxplots:

```
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

pd.set_option("display.max_rows", None, "display.max_columns", None)

data['outlier'] = (((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1))

print(data[data.outlier == True])
```

Tem-se, acima, os valores que são considerados *outliers* (238 observações em 506), pelo método do boxplot, para os dados do *dataset* BostonHousing. Ressalta-se uma vez mais que essa análise é simplista e preliminar, pois, dentre outras coisas, desconsidera-se o fato de se ter dados multivariados ao analisá-los individualmente, atributo a atributo.

A seguir, identifica-se os *outliers* visualmente em um gráfico para verificar se os registros encontrados são, de fato, *outliers* no scatterplot. Vale ressaltar que as variáveis utilizadas foram escolhidas arbitrariamente.

```
plt.figure(figsize=(20,10))

plt.scatter(x = data.lstat, y = data.medv, s = 50,
            c = data.outlier, cmap = 'viridis')

plt.xlabel('lstat')
plt.ylabel('medv')

plt.show()
```

Como é possível observar, os *outliers* (representados na cor amarela) não são visualmente condizentes com o esperado (esperava-se que os mesmos se encontrassem às margens do gráfico). Possivelmente isto se deve ao fato de se estar analisando os *outliers* unidimensionalmente (atributo a atributo) em dados que são, na verdade, multivariados.
